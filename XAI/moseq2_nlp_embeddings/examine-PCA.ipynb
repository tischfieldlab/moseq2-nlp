```python
import pandas as pd
import numpy as np
import seaborn as sns
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

# rOT

```python
model0 = Doc2Vec.load('rOT/doc2vec.0.model')
model1 = Doc2Vec.load('rOT/doc2vec.1.model')
print(model0)
print(model1)
```

```python
labels = pd.read_csv('labels.csv').to_numpy()
classes = np.unique(labels)
# colors = {'24hrs meloxicam':'r', '24hrs saline':'g', '4hrs carrageenan':'b', 'baseline':'c', 'baseline meloxicam':'k'}
colors = {'24hrs meloxicam':"#58d68d", '24hrs saline':"#f7dc6f", '4hrs carrageenan':"#e59866", 'baseline':"#5d6d7e", 'baseline meloxicam':"#d2b4de"}


meloxicam24h = mpatches.Patch(color="#58d68d", label='24hrs meloxicam')
saline24h = mpatches.Patch(color="#f7dc6f", label='24hrs saline')
carrageenan = mpatches.Patch(color="#e59866", label='4hrs carrageenan')
baseline = mpatches.Patch(color="#5d6d7e", label='baseline')
baselineMeloxicam =  mpatches.Patch(color="#d2b4de", label='baseline meloxicam')
```

```python
c = []
for label in labels:
    if label[0] in colors:
        c.append(colors[label[0]])
```

```python
m0_features = pd.DataFrame(model0.dv.vectors)
m1_features = pd.DataFrame(model1.dv.vectors)

pca1 = PCA(n_components=2)
pca1.fit(m0_features)
pca_f1 = pca1.transform(m0_features)

pca2 = PCA(n_components=2)
pca2.fit(m0_features)
pca_f2 = pca2.transform(m1_features)

fig = plt.figure(figsize=(12,5))
ax1 = fig.add_subplot(1,2,1)


ax1.scatter(pca_f1[:, 0], pca_f1[:, 1], s=75,c=c, cmap=plt.cm.nipy_spectral)


ax2 = fig.add_subplot(1,2,2)
ax2.scatter(pca_f2[:, 0], pca_f2[:, 1], s=75,c=c, cmap=plt.cm.nipy_spectral)

fig.tight_layout(h_pad=2)
ax1.set_title('Model 0')
ax2.set_title('Model 1')
title = fig.suptitle('Embeddings PCA rOT 2D', size='16')
fig.subplots_adjust(top=0.88)
legend = fig.legend(handles=[meloxicam24h,saline24h,carrageenan,baseline,baselineMeloxicam],bbox_to_anchor=(1.04,0.5), loc="center left")
fig.savefig('Embeddings_PCA_rOT_2D.png',bbox_extra_artists=(legend,title), bbox_inches='tight')

# fig.savefig('embeddings_PCA_RoT_2D.png')
```

```python
m0_features = pd.DataFrame(model0.dv.vectors)
m1_features = pd.DataFrame(model1.dv.vectors)

pca1 = PCA(n_components=3)
pca1.fit(m0_features)
pca_f1 = pca1.transform(m0_features)

pca2 = PCA(n_components=3)
pca2.fit(m0_features)
pca_f2 = pca2.transform(m1_features)

fig = plt.figure(figsize=(12,5))
ax1 = fig.add_subplot(1,2,1, projection='3d')


ax1.scatter(pca_f1[:, 0], pca_f1[:, 1],  pca_f1[:, 2],s=75,c=c, cmap=plt.cm.nipy_spectral)

ax2 = fig.add_subplot(1,2,2, projection='3d')
ax2.scatter(pca_f2[:, 0], pca_f2[:, 1], pca_f2[:, 2],s=75, c=c, cmap=plt.cm.nipy_spectral)

fig.tight_layout(h_pad=2)
ax1.set_title('Model 0')
ax2.set_title('Model 1')
title = fig.suptitle('Embeddings PCA rOT 3D', size='16')
fig.subplots_adjust(top=0.88)
legend = fig.legend(handles=[meloxicam24h,saline24h,carrageenan,baseline,baselineMeloxicam],bbox_to_anchor=(1.04,0.5), loc="center left")
fig.savefig('Embeddings_PCA_rOT_3D.png',bbox_extra_artists=(legend,title), bbox_inches='tight')
```
